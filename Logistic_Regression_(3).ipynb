# Logistic Regression

1. Linear Model of classification, assumes linear relationship between feature & target
2. y = e^(b0 + b1x) / (1 + e^(b0 + b1x))
3. Returns class probabilities
4. Hyperparameter : C - regularization coef
5. Fundamentally suited for bi-class classification

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
import numpy as np
```

```python
X,y = make_blobs(n_features=2, n_samples=1000, cluster_std=2,centers=2)
```

```python
plt.scatter(X[:,0],X[:,1],c=y,s=10)
```

```python
h = .02
x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X,y)
```

```python
Z = lr.predict(np.c_[xx.ravel(), yy.ravel()])
```

```python
Z = Z.reshape(xx.shape)
plt.pcolormesh(xx, yy, Z, cmap=plt.cm.Paired)
plt.scatter(X[:,0],X[:,1],c=y,s=10)
```
